```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url = "https://raw.githubusercontent.com/phinum/IME/main/total_returns.csv" 
df = pd.read_csv(url)
df = df.drop('Unnamed: 0', axis = 1)
df = df.drop('Date', axis = 1)
df
```

```python
df1 = df.transpose()
df1.shape
df1.reset_index(drop=True, inplace=True)

df2 = df1.transpose()
df2_t = df2.transpose()
df2_t
```

```python
cov = np.cov(df2_t)
cov
```

```python
np.shape(cov)
```

```python
mean = df2.mean().values
```

```python
mean
```

```python
import numpy as np
from sklearn.covariance import EmpiricalCovariance
from sklearn.datasets import make_gaussian_quantiles
real_cov = np.cov(df2_t)
rng = np.random.RandomState(0)
X = rng.multivariate_normal(mean = mean, cov= real_cov, size= 500)
cov = EmpiricalCovariance().fit(X)
```
