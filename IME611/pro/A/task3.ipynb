```python
import pandas as pd
import numpy as np
import seaborn as sns
%matplotlib inline
```

```python
df1 = pd.read_csv('nifty50.csv')
df1 = df1.drop(df1.tail(50).index)
df1
```

```python
df1 = df1.dropna(axis=1)
df1 = df1.drop('Date', axis = 1)
name = []
for i in range(len(df1.columns)):
    name.append(df1.columns[i])
```

```python
df1
```

```python
a = [0, 245, 246, 493, 494, 739, 740, 981, 982, 1231]
```

```python
df = pd.DataFrame()
for i in range(len(name)):
    erm = []
    erm.append((df1[name[i]][a[1]] - df1[name[i]][a[0]])/df1[name[i]][a[0]])
    erm.append((df1[name[i]][a[3]] - df1[name[i]][a[2]])/df1[name[i]][a[2]])
    erm.append((df1[name[i]][a[5]] - df1[name[i]][a[4]])/df1[name[i]][a[4]])
    erm.append((df1[name[i]][a[7]] - df1[name[i]][a[6]])/df1[name[i]][a[6]])
    erm.append((df1[name[i]][a[9]] - df1[name[i]][a[8]])/df1[name[i]][a[8]])
    df[name[i]] = erm
df
```

```python
df.to_csv('chomti.csv')
```

```python
print(df.mean())
print(df1.std())
print(df1.var())
```
