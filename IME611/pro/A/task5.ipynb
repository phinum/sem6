```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as pdr
import datetime as dt
```

```python
df = pd.read_csv('n50.csv')
df = df.drop('Date', axis = 1)
```

```python
a = [0, 243, 244, 491, 492, 736, 737, 977]
```

```python
erm = []
erm.append((df['Close'][a[1]] - df['Close'][a[0]])/df['Close'][a[0]])
erm.append((df['Close'][a[3]] - df['Close'][a[2]])/df['Close'][a[2]])
erm.append((df['Close'][a[5]] - df['Close'][a[4]])/df['Close'][a[4]])
erm.append((df['Close'][a[7]] - df['Close'][a[6]])/df['Close'][a[6]])
```

```python
Erm = np.mean(erm)
```

```python
rf = 0.05
```

```python
def abline(slope, intercept):
    axes = plt.gca()
    x_vals = np.linspace(0,0.5,10)
    y_vals = intercept + slope * x_vals
    plt.plot(x_vals, y_vals, '--')
```

```python
abline(Erm-rf, rf)
```

```python
df = df.pct_change()
df = df.dropna()
df
```

```python
df1 = pd.read_csv('nifty50.csv')
df1 = df1.dropna(axis=1)
df1 = df1.drop('Date', axis = 1)
name = []
for i in range(len(df1.columns)):
    name.append(df1.columns[i])
```

```python
df1 = df1.pct_change()
df1 = df1.dropna()
df1 = df1.drop(df1.tail(303).index)
df1
```

```python
beta = []
for i in range(48):
    x = np.cov(df1[name[i]],df['Close'])[0][1]/np.var(df['Close'])
    beta.append(x)
    
```

```python
df2 = pd.read_csv('nifty50.csv')
df2 = df2.dropna(axis=1)
df2 = df2.drop('Date', axis = 1)
name = []
for i in range(len(df2.columns)):
    name.append(df2.columns[i])
    
df2 = df2.drop(df2.tail(303).index)
df2
```

```python
ERI = []

for i in range(48):
    eri = []
    eri.append((df2[name[i]][a[1]] - df2[name[i]][a[0]])/df2[name[i]][a[0]])
    eri.append((df2[name[i]][a[3]] - df2[name[i]][a[2]])/df2[name[i]][a[2]])
    eri.append((df2[name[i]][a[5]] - df2[name[i]][a[4]])/df2[name[i]][a[4]])
    eri.append((df2[name[i]][a[7]] - df2[name[i]][a[6]])/df2[name[i]][a[6]])
    x = np.mean(eri)
    ERI.append(x)
```

```python
idx = []
for i in range(len(beta)):
    if beta[i] < 0:
        beta.pop(i)
        ERI.pop(i)
        idx.append(i)
        i = i + 1
```

```python
plt.figure(figsize=(15,8))
abline(Erm-rf, rf)
plt.scatter(beta, ERI)
plt.xlabel('beta')
plt.ylabel('return')
plt.show()
```
