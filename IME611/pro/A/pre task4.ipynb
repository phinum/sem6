```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
import scipy.stats
```

```python
df = pd.read_csv('before_COVID.csv')
```

```python
a = []

for i in range(len(df.columns)-2):
    a.append(df.columns[i+2])
```

```python
df2 = pd.DataFrame()

for i in range(len(a)):
    b = []
    b.append(np.nan)
    for j in range(int(df.shape[0]/5)):
        b.append((df[a[i]][j+5] - df[a[i]][j])/df[a[i]][j])
        j = j + 5
    df2[a[i]] = b
    
df2 = df2.iloc[1:]
df2
```

```python
b = []

for i in range(len(a)):
    st = stats.shapiro(df2[a[i]])
    b.append(st.pvalue)
    
b
```

```python
others = []   

for i in range(len(a)):
    percentile_bins = np.linspace(0,100,51)
    percentile_cutoffs = np.percentile(df2[a[i]], percentile_bins)
    observed_frequency, bins = (np.histogram(df2[a[i]], bins=percentile_cutoffs))
    cum_observed_frequency = np.cumsum(observed_frequency)

    dist_names = ['weibull_min','norm','weibull_max','beta','invgauss','uniform','gamma','expon','lognorm','pearson3',
                  'triang']

    chi_square_statistics = []

    for distribution in dist_names:
        dist = getattr(scipy.stats, distribution)
        param = dist.fit(df2[a[i]])
        print("{}\n{}\n".format(dist, param))

        cdf_fitted = dist.cdf(percentile_cutoffs, *param)
        expected_frequency = []
        for bin in range(len(percentile_bins)-1):
            expected_cdf_area = cdf_fitted[bin+1] - cdf_fitted[bin]
            expected_frequency.append(expected_cdf_area)

        # Chi-square Statistics
        expected_frequency = np.array(expected_frequency) * 201
        cum_expected_frequency = np.cumsum(expected_frequency)
        ss = sum (((cum_expected_frequency - cum_observed_frequency) ** 2) / cum_observed_frequency)
        chi_square_statistics.append(ss)

    results = pd.DataFrame()
    results['Distribution'] = dist_names
    results['chi_square'] = chi_square_statistics
    x = results['chi_square'].idxmin()
    results.sort_values(['chi_square'], inplace=True)
    print(results)
    others.append(results['Distribution'][x])
```
