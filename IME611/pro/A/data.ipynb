```python
import pandas as pd
import pandas_datareader.data as pdr
import datetime as dt
```

```python
fin = dt.datetime(2021, 3, 15)
ini = dt.datetime(2016, 1, 1)
```

```python
df = pd.read_csv('MW-NIFTY-50-22-Mar-2021.csv')
ticker = []
for i in range(1,51):
    ticker.append(df["SYMBOL \n"][i] + '.NS')
    
len(ticker)
```

```python
df1 = pd.DataFrame()

for i in range(len(ticker)):
    try:
        o = pdr.get_data_yahoo(ticker[i],ini,fin)
        df1[str(ticker[i])] = o["Close"]
    except:
        print('error')
        print(i)

df1
```

```python
df1.to_csv('nifty50.csv', index = True)
```
