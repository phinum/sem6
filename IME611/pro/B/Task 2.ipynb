```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader.data as pdr
import datetime as dt
from statistics import stdev
```

```python
ticker = ['ADANIPORTS.NS', 'BRITANNIA.NS', 'TCS.NS', 'HINDALCO.NS', 'HINDUNILVR.NS', 'TATASTEEL.NS', 'JSWSTEEL.NS', 
         'DIVISLAB.NS', 'NESTLEIND.NS', 'ASIANPAINT.NS', 'TITAN.NS', 'BAJAJFINSV.NS', 'RELIANCE.NS', 'BAJFINANCE.NS',
         'HDFCBANK.NS', ]

no_of_stocks = [54, 4, 374, 216, 565, 249, 2096, 188, 62, 67, 869, 28, 563, 232, 32]
```

```python
df = pd.DataFrame()
df['Stock_names'] = ticker
df['no_of_stocks'] = no_of_stocks
df
```

```python
fin = dt.datetime(2021, 5, 8)
ini = dt.datetime(2021, 4, 7)
```

```python
df1 = pd.DataFrame()

for i in range(len(ticker)):
    try:
        o = pdr.get_data_yahoo(ticker[i],ini,fin)
        df1[str(ticker[i])] = o["Close"]
    except:
        print('error')
        print(i)

o = pdr.get_data_yahoo('^NSEI',ini,fin)
df1['Nifty50'] = o["Close"]
```

```python
df1.shape[0]
```

```python
x = []

for j in range(df1.shape[0]):
    initial_value = 0
    for i in range(len(ticker)):
        initial_value += df1[ticker[i]][j]*no_of_stocks[i] 

    x.append(initial_value)
    
df1['Portfolio'] = x
df1
```

### Total Returns during trading period

```python
Return_P = ((df1['Portfolio'][20] - df1['Portfolio'][0]) / df1['Portfolio'][0])*100
Return_B = ((df1['Nifty50'][20] - df1['Nifty50'][0]) / df1['Nifty50'][0])*100

print('Total Return on Portfolio in %: ', Return_P, '%')
print('Total Return on Benchmark in %: ', Return_B, '%')
```

### Volatility

```python
x = df1["Portfolio"].pct_change()
vol_P = x.std()*np.sqrt(21)*100
x1 = df1["Nifty50"].pct_change()
vol_B = x1.std()*np.sqrt(21)*100

print('Volatility of Portfolio in %: ', vol_P, '%')
print('Volatility of Benchmark in %: ', vol_B, '%')
```

### Sharpe Ratio

```python
rf = 0.06
rfp = ((1+rf)**(30/365) -1)*100

sr = (Return_P - rfp)/vol_P
print('Sharpe Ratio of Portfolio: ', sr)
```

### Sortino Ratio

```python
x = df1["Portfolio"].pct_change()
y = []
for i in range(len(x)):
    if x[i] < 0:
        y.append(x[i])
print(y)
neg_vol = stdev(y)*np.sqrt(21)*100

sor = (Return_P - rfp)/neg_vol
print('Sortino Ratio of Portfolio: ', sor)
```

### Maximum Draw-down

```python
df1["daily_ret_P"] = df1["Portfolio"].pct_change()
df1["cum_return"] = (1 + df1["daily_ret_P"]).cumprod()
df1["cum_roll_max"] = df1["cum_return"].cummax()
df1["drawdown"] = df1["cum_roll_max"] - df1["cum_return"]
df1["drawdown_pct"] = df1["drawdown"]/df1["cum_roll_max"]
max_dd = df1["drawdown_pct"].max()
print('Maximum Draw-down of Portfolio in %: ', max_dd*100, '%')
```

### Calmar Ratio

```python
cr = Return_P/(max_dd*100)
print('Calmar Ratio of Portfolio: ', cr)
```

### Beta of the Portfolio

```python
df1['daily_ret_B'] = df1["Nifty50"].pct_change()
p = df1['daily_ret_P'].dropna()
q = df1['daily_ret_B'].dropna()
beta = np.cov(p,q)[0][1]/np.var(q)
print('Beta of Portfolio: ', beta)
```

### Alpha Generated

```python
alpha = Return_P - rfp + beta*(Return_B - rfp)
print('Alpha of Portfolio in %: ', alpha, '%')
```
